```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data/pingpong_match.csv',encoding='gbk')
data
```

```python
data.columns = data.iloc[0]
data = data.drop(data.index[0])
data = data.reset_index(drop=True).astype(float)
data
```

```python
data['elo1'] = 1792.00
data['elo2'] = 1553.00

K = 40

def update_elo(elo_winner, elo_loser, K=40):
    expected_win = 1 / (1 + 10 ** ((elo_loser - elo_winner) / 400))
    update_winner = elo_winner + K * (1 - expected_win)
    update_loser = elo_loser - K * expected_win
    return update_winner, update_loser


for index, row in data.iterrows():
    if index == 0: 
        continue
    
    point_victor = row['point_victor']
    
    prev_elo1 = data.at[index - 1, 'elo1']
    prev_elo2 = data.at[index - 1, 'elo2']

    if point_victor == 1:
        updated_elo1, updated_elo2 = update_elo(prev_elo1, prev_elo2, K)
    else: 
        updated_elo2, updated_elo1 = update_elo(prev_elo2, prev_elo1, K)

    data.at[index, 'elo1'] = updated_elo1
    data.at[index, 'elo2'] = updated_elo2

data
```
