```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 击球分布

```python
data = pd.read_csv('Wimbledon_featured_matches.csv')
data
```

```python
data1 = data[['match_id', 'player1', 'player2', 'set_no', 'game_no', 'point_no', \
             'server','serve_no', 'speed_mph', 'serve_width','serve_depth', 'return_depth']]
data1
```

```python
all_players = np.union1d(data['player1'].unique(), data['player2'].unique())
all_players
```

```python
def speed_score(speed):
    if speed < 110:
        return 1
    elif 110 <= speed < 130:
        return 2
    else:
        return 3

def serve_width_score(serve_width):

    return {'W': 3, 'B': 1, 'C': 3, 'BC': 2, 'BW': 2}.get(serve_width, 0)

def serve_depth_score(serve_depth):

    return {'CLT': 2, 'NCTL': 1}.get(serve_depth, 0)

def return_depth_score(return_depth):
    
    return {'D': 2, 'ND': 1}.get(return_depth, 0)
```

```python
def calculate_performance_score(player, set_no, game_no):
    filtered_df = data[((data['set_no'] == set_no) & (data['game_no'] == game_no)) &
                        ((data['player1'] == player) | (data['player2'] == player))]
    
    scores = []
    for index, row in filtered_df.iterrows():
        score = 0
        if (row['server'] == 1 and row['player1'] == player) or (row['server'] == 2 and row['player2'] == player):
            score += speed_score(row['speed_mph'])
            score += serve_width_score(row['serve_width'])
            score += serve_depth_score(row['serve_depth'])
        else:
            score += return_depth_score(row['return_depth'])
        
        scores.append(score)
    
    if scores:
        return sum(scores) / len(scores)
    else:
        return None 
```

```python
def calculate_row_performance_score(row, data):

    player1_score = calculate_performance_score(row['player1'], row['set_no'], row['game_no'])

    player2_score = calculate_performance_score(row['player2'], row['set_no'], row['game_no'])
    
    return player1_score, player2_score
```

```python
performance_scores = {}

for player in all_players:
    performance_scores[player] = {}
    unique_set_games = data[['set_no', 'game_no']].drop_duplicates()
    
    for _, set_game in unique_set_games.iterrows():
        set_no = set_game['set_no']
        game_no = set_game['game_no']
        
        score = calculate_performance_score(player, set_no, game_no)
        performance_scores[player][f"{set_no}-{game_no}"] = score

def get_quality_score(player, set_game, performance_scores):
    return performance_scores.get(player, {}).get(set_game, None)

data['player1_quality'] = data.apply(lambda row: get_quality_score(row['player1'], f"{row['set_no']}-{row['game_no']}", performance_scores), axis=1)

data['player2_quality'] = data.apply(lambda row: get_quality_score(row['player2'], f"{row['set_no']}-{row['game_no']}", performance_scores), axis=1)

data
```

```python
for player in all_players:
    set_games = []
    qualities = []
    for set_no in sorted(data['set_no'].unique()):
        for game_no in sorted(data[data['set_no'] == set_no]['game_no'].unique()):
            score = calculate_performance_score(player, set_no, game_no)
            if score is not None:
                set_games.append(f"{set_no}-{game_no}")
                qualities.append(score)
    
    if set_games:
        plt.figure(figsize=(10, 6))
        plt.plot(set_games, qualities, marker='o', linestyle='-', label=f'Quality of {player}')
        plt.title(f'Performance Score of {player} Across Sets and Games')
        plt.xlabel('Set-Game')
        plt.ylabel('Quality Score')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
```

```python
data.to_csv('data/quality_data.csv')
```

```python
filtered_data = data[data['game_victor'] != 0].copy()
filtered_data['predicted_victor'] = np.where(filtered_data['player1_quality'] > filtered_data['player2_quality'], 1, 2)
actual_victor = filtered_data['game_victor']
predicted_victor = filtered_data['predicted_victor']

filtered_data['correct_prediction'] = filtered_data['predicted_victor'] == filtered_data['game_victor']

correct_predictions = filtered_data['correct_prediction'].sum()
total_predictions = len(filtered_data)

accuracy = correct_predictions / total_predictions

print(f"预测正确率为: {accuracy:.2%}")
```
